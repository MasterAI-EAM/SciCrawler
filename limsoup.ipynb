```python
from LimeSoup import ElsevierSoup
import json

def store_json(article, input_d, output_d):
    with open(input_d+article, 'r', encoding = 'utf-8') as f:
        html_str = f.read()
        try:
            data = ElsevierSoup.parse(html_str)
        except:
            return False
    article = article.replace('.xml', '')
    name = output_d+article+'.json'
    if not os.path.exists(name):
        with open(name, 'w', encoding = 'utf-8') as f:
            json.dump(data, f, sort_keys=True, indent=4, ensure_ascii=False)
```

# Springer

```python
from chemdataextractor.doc import Document, Title, Heading, Paragraph, Citation, Table, Figure, Caption, Footnote
from chemdataextractor.reader import NlmXmlReader
# from chemdataextractor.reader.markup import XmlReader
from io import BytesIO


def store_xml2json(filename):
    store_json = {"Sections":[]}
    f = open(filename, 'rb')
    print(type(f))
    data = f.read()
    print(type(data))
    doc = Document.from_file(f, readers=[NlmXmlReader()])
    tmp = {}
    for e in doc.elements:
        if type(e) == Heading:
            if tmp != {}:
                store_json["Sections"].append(tmp)
            tmp = {}
            tmp["name"] = str(e)
            tmp["content"]=[]
        if type(e) == Paragraph:
            word_num = len(str(e).split(' '))
            if word_num > 100:
                str_e = str(e).replace('\n', '')
                if str_e not in tmp["content"]:
                    tmp["content"].append(str_e)
                    
    return store_json
```

```python
import os
from tqdm import tqdm
import json

names = os.listdir('../springer_xml/energy&material/')
for n in tqdm(names[5:]):
    sj = store_xml2json('../springer_xml/energy&material/'+n)
    json_str = json.dumps(sj, indent=4)
    nn = n.replace('.xml', '.json')
    with open('../springer_json/'+nn, 'w', encoding='utf-8') as json_file:
        json_file.write(json_str)
```
